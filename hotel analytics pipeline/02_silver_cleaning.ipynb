```python
from pyspark.sql.functions import col, to_date

spark.sql("CREATE DATABASE IF NOT EXISTS hotel_silver")
df = spark.table("hotel_bronze.bookings_raw")

df_clean = (
    df
    .filter(col("adr") > 0)
    .withColumn("reservation_status_date", to_date(col("reservation_status_date")))
    .withColumn("revenue", col("adr") * col("stays_in_weekend_nights"))
)

df_clean.write.format("delta") \
  .mode("overwrite") \
  .saveAsTable("hotel_silver.bookings_clean")
```
